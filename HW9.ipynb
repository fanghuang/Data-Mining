```python
def AttributeEncoder (m):
    le.fit(m)
    return le.transform(m)
```

```python
Ti.Class=AttributeEncoder(Ti.Class)
Ti.Sex=AttributeEncoder(Ti.Sex)
Ti.Age=AttributeEncoder(Ti.Age)
Ti.Survived=AttributeEncoder(Ti.Survived)
```

```python
weights=1.0*Ti.Freq/Ti.Freq.sum()
weights
```

```python
clf = tree.DecisionTreeClassifier(max_depth=2)
train=Ti.drop(['Survived','Freq'],axis=1)
```

```python
train
```

```python
clf=clf.fit(train,Ti.Survived,sample_weight=weights)
```

```python
from sklearn.externals.six import StringIO
with open("Titanic.dot",'w') as f:
    f = tree.export_graphviz(clf,out_file=f)
```
